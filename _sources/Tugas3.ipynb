##**penyelesaian sistem persamaan linier tentang matrix invers**

Invers matriks adalah kebalikan (invers) dari sebuah matriks yang apabila matriks tersebut dikalikan dengan inversnya, akan menjadi matriks identitas. Invers matriks dilambangkan dengan A-1. Suatu matriks dikatakan memiliki invers jika determinan dari matriks tersebut tidak sama dengan nol.



Untuk menentukan invers dari sebuah matriks, terdapat dua aturan berdasarkan ordonya, yaitu ordo 2x2 dan ordo 3x3.

Untuk mencari invers dari matriks berordo 2x2, dapat menggunakan rumus berikut:

Jika diberikan matriks:

\begin{equation*}
A=
\begin{bmatrix}
 a & b \\ c & d \end{bmatrix}
\end{equation*}


Maka invers dari matriks A, dilambangkan sebagai \(A^{-1}\), dapat dihitung menggunakan rumus:

\begin{equation*}
A^{-1} = \frac{1}{ad - bc}
\begin{bmatrix}
 d & -b \\ -c & a \end{bmatrix}
\end{equation*}

Untuk mencari invers dari matriks berordo 3x3, kita dapat menggunakan metode eliminasi Gauss-Jordan. Proses ini melibatkan transformasi matriks awal menjadi matriks identitas dengan mengaplikasikan serangkaian operasi baris dan operasi kolom. Setelah matriks awal menjadi matriks identitas, matriks yang terkait dengan matriks identitas tersebut adalah invers dari matriks awal.

Langkah-langkahnya adalah sebagai berikut:

Misalkan kita memiliki matriks 3x3 A:

\begin{equation*}
A=
\begin{bmatrix}
  a & b & c \\ d & e & f \\ g & h & i \end{bmatrix}
\end{equation*}

Kita juga buat matriks identitas 3x3, disini dilambangkan sebagai \(I\):

\begin{equation*}
I=
\begin{bmatrix}
 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}
\end{equation*}

Kemudian kita gabungkan matriks A dengan matriks identitas I, membentuk matriks augmented \(\):

\begin{equation*}
[A | I] =
\begin{bmatrix}
a & b & c & | & 1 & 0 & 0 \\ d & e & f & | & 0 & 1 & 0 \\ g & h & i & | & 0 & 0 & 1 \end{bmatrix}
\end{equation*}

Lalu, kita terapkan serangkaian operasi baris dan operasi kolom untuk mereduksi bagian kiri (matriks A) menjadi matriks identitas. Setelah matriks A menjadi identitas, bagian kanan matriks augmented adalah invers dari matriks awal.



###**sifat sifat matrix invers**

Invers dari sebuah matriks memiliki beberapa sifat yang penting dalam aljabar matriks. Berikut adalah beberapa sifat penting dari invers matriks:

1. **Kepemilikan**: Tidak semua matriks memiliki invers. Matriks yang memiliki invers disebut matriks nonsingular atau matriks invertibel. Matriks yang tidak memiliki invers disebut matriks singular.

2. **Unik**: Jika sebuah matriks memiliki invers, maka inversnya adalah unik. Artinya, tidak ada dua matriks yang berbeda yang memiliki invers yang sama.

3. **Produk dengan invers**: Jika A adalah matriks yang memiliki invers, maka perkalian matriks A dengan inversnya, baik dari kiri maupun dari kanan,
\begin{equation*}
A \times A^{-1} = A^{-1} \times A = I
\end{equation*}
menghasilkan matriks identitas. Simbolisnya, , di mana I adalah matriks identitas yang sesuai dengan ukuran matriks A.

4. **Invers dari invers**: Invers dari invers dari suatu matriks adalah matriks itu sendiri. Dengan kata lain, jika A adalah matriks yang memiliki invers, maka
\begin{equation*}
(A^{-1})^{-1} = A
\end{equation*}

5. **Invers dari matriks transpos**: Invers dari transpos suatu matriks adalah transpos dari invers matriks tersebut. Dalam simbol, jika A adalah matriks yang memiliki invers, maka
\begin{equation*}
(A^T)^{-1} = (A^{-1})^T
\end{equation*}

6. **Invers dari perkalian matriks**: Jika A dan B adalah matriks yang memiliki invers, maka perkalian matriks AB juga memiliki invers, dan invers dari AB adalah hasil perkalian inversnya dalam urutan yang terbalik. Simbolisnya, jika \(A\) dan \(B\) adalah matriks yang memiliki invers, maka
\begin{equation*}
(AB)^{-1} = B^{-1}A^{-1}
\end{equation*}

7. **Invers dari matriks diagonal**: Invers dari matriks diagonal dapat diperoleh dengan mengambil invers dari setiap elemen diagonalnya. Jika sebuah elemen diagonal adalah nol, maka matriks tidak memiliki invers.

8. **Invers dari matriks identitas**: Invers dari matriks identitas adalah matriks identitas itu sendiri.




###**mencari invers dengan eliminasi geus jordan**


````python
import numpy as np

def cek_invers(matrix):
    # Cek apakah matriks adalah matriks persegi
    if matrix.shape[0] != matrix.shape[1]:
        print("Matriks bukan matriks persegi, tidak memiliki invers.")
        return False

    # Cek determinan matriks
    det = np.linalg.det(matrix)

    # Jika determinan tidak nol, maka matriks memiliki invers
    if det != 0:
        print("Matriks memiliki invers.")
        return True
    else:
        print("Matriks tidak memiliki invers.")
        return False

# Contoh penggunaan
A = np.array([[2, 1, 3],
              [1, 0, 2],
               [3, 2, 1]])

B = np.array([[1, 2, 3],
              [2, 4, 6],
              [3, 6, 9]])

print("Matriks A:")
print(A)
cek_invers(A)

print("\nMatriks B:")
print(B)
cek_invers(B)
````

###**solusi sisitem persamaan linier dengan matrix invers**

Untuk menyelesaikan sebuah sistem persamaan linear dengan menggunakan matriks invers, langkah-langkah umumnya adalah sebagai berikut:

Misalkan kita memiliki sebuah sistem persamaan linear:
\[ Ax = b \]

Di mana:
- \( A \) adalah matriks koefisien yang memiliki invers,
- \( x \) adalah vektor variabel yang ingin kita cari,
- \( b \) adalah vektor hasil.

Langkah-langkahnya adalah sebagai berikut:

1. **Cari invers dari matriks koefisien**: Jika matriks koefisien \( A \) memiliki invers (\( A^{-1} \)), kita dapat mengalikan kedua sisi persamaan dengan invers dari matriks koefisien tersebut:
\begin{equation*}A^{-1}Ax = A^{-1}b\end{equation*}

2. **Perkalian invers dengan \( A \) pada kedua sisi**: Karena \( A^{-1}A \) adalah matriks identitas, kita dapat menyederhanakan persamaan menjadi:
\begin{equation*}Ix = A^{-1}b\end{equation*}


3. **Hitung hasil perkalian**: Karena ( Ix = x \) untuk setiap vektor \( x \), hasil akhirnya adalah:
\begin{equation*}x = A^{-1}b\end{equation*}


Dengan demikian, kita dapat menemukan solusi vektor variabel \( x \) dengan mengalikan invers dari matriks koefisien dengan vektor hasil. Namun, perlu diingat bahwa metode ini hanya dapat digunakan jika matriks koefisien \( A \) memiliki invers.

Berikut adalah contoh kode Python menggunakan NumPy untuk menyelesaikan sistem persamaan linear dengan matriks invers:

```python
import numpy as np

# Matriks koefisien
A = np.array([[2, 1, -1],
              [1, -1, 1],
              [1, 1, 1]])

# Vektor hasil
b = np.array([4, 2, 6])

# Mencari invers matriks koefisien
A_inv = np.linalg.inv(A)

# Menyelesaikan sistem persamaan linear dengan matriks invers
x = np.dot(A_inv, b)

print("Solusi vektor x:", x)
```

Perhatikan bahwa langkah ini hanya berlaku jika matriks koefisien memiliki invers. Jika matriks tersebut singular atau tidak memiliki invers, Anda harus menggunakan metode lain untuk menyelesaikan sistem persamaan linear tersebut.

````python
import numpy as np

# Matriks koefisien
A = np.array([[2, 1, -1],
              [1, -1, 1],
              [1, 1, 1]])

# Vektor hasil
b = np.array([4, 2, 6])

# Mencari invers matriks koefisien
A_inv = np.linalg.inv(A)

# Menyelesaikan sistem persamaan linear dengan matriks invers
x = np.dot(A_inv, b)

print("Solusi vektor x:", x)
````

4. **Kembalikan Matriks**:
   Setelah langkah-langkah di atas selesai dilakukan, matriks akan dimodifikasi sesuai dengan prosedur Gauss-Jordan. Matriks yang telah dimodifikasi akan dikembalikan sebagai hasil.

Ini adalah langkah-langkah yang dilakukan untuk mengimplementasikan metode eliminasi Gauss-Jordan secara langsung, yang menghasilkan invers matriks.

````python
import numpy as np

def eliminasi_gauss_jordan(matriks):  #Fungsi untuk melakukan eliminasi Gauss-Jordan pada sebuah matriks.
    matriks = np.array(matriks, dtype=float)  # Konversi matriks menjadi numpy array
    n = len(matriks)

    for i in range(n):
        # Pivoting
        if matriks[i][i] == 0:
            for j in range(i + 1, n):
                if matriks[j][i] != 0:
                    matriks[i], matriks[j] = matriks[j].copy(), matriks[i].copy()
                    break

        # Normalisasi
        pembagi = matriks[i][i]
        matriks[i] /= pembagi

        # Eliminasi
        for j in range(n):
            if i != j:
                pengali = matriks[j][i]
                matriks[j] -= pengali * matriks[i]

    return matriks

def invers(matriks): #Fungsi untuk mencari invers dari sebuah matriks menggunakan eliminasi Gauss-Jordan.
    matriks = np.array(matriks, dtype=float)  # Konversi matriks menjadi numpy array
    n = len(matriks)
    matriks_identitas = np.identity(n)
    matriks_augment = np.concatenate((matriks, matriks_identitas), axis=1)
    matriks_reduksi = eliminasi_gauss_jordan(matriks_augment)

    invers_matriks = matriks_reduksi[:, n:]  # Ambil bagian matriks kanan setelah kolom terakhir
    return invers_matriks

# Contoh penggunaan:
A = [[1, 2, 3], [0, 1, 4], [5, 6, 0]]  # Matriks input
A_invers = invers(A)
print("Invers dari A:")
print(A_invers)
````

````python
import numpy as np

def gauss_jordan(a, b):
    a = np.array(a, dtype=float)
    b = np.array(b, dtype=float)

    n = len(b)
    ab = np.column_stack((a, b))  #melakukan fungsi inisialisasi

    #membuat aturan mengecek garis diagonal pada matrix
    for k in range(n):
        if ab[k, k] == 0:
            for i in range(k + 1, n):
                if ab[i, k] != 0:
                    ab[[k, i]] = ab[[i, k]]
                    break

        pivot = ab[k, k]
        ab[k] /= pivot

        #melakukan eliminasi utk menghilangkan elemen di kolom yang sama dengan elemen diagonal utama
        for i in range(n):
            if i != k:
                factor = ab[i, k]
                ab[i] -= factor * ab[k]

    x = ab[:, -1]
    a_transformed = ab[:, :-1]
    return x, a_transformed

a = [[2, 1, -1],
     [1, -1, 1],
     [1, 1, 1]]
b = [[2], [2], [2]]

solution, transformed_a = gauss_jordan(a, b)

print("Hasil nilai matrix diagonal :")
print(transformed_a)
print("Hasil nilai vektor :")
print(solution)
````
