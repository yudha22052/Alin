```python
def minor(matrix, row, col):
    """Menghasilkan minor dari elemen matriks pada baris row dan kolom col"""
    return [row[:col] + row[col+1:] for row in (matrix[:row] + matrix[row+1:])]

def determinant(matrix):
    """Menghitung determinan dari matriks secara rekursif"""
    if len(matrix) == 2:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]
    det = 0
    for i in range(len(matrix)):
        det += ((-1) ** i) * matrix[0][i] * determinant(minor(matrix, 0, i))
    return det

def cofactor(matrix, i, j):
    """Menghasilkan kofaktor dari elemen matriks pada baris i dan kolom j"""
    minor_matrix = minor(matrix, i, j)
    sign = (-1) ** (i + j)
    minor_det = determinant(minor_matrix)
    return sign * minor_det

def transpose(matrix):
    """Menghasilkan transpose dari matriks"""
    return [[matrix[j][i] for j in range(len(matrix))] for i in range(len(matrix[0]))]

def matrix_inverse(matrix):
    """Menghasilkan invers dari matriks menggunakan metode adjoin"""
    det = determinant(matrix)
    if det == 0:
        raise ValueError("Matriks tidak memiliki invers karena determinannya nol")
    adjoint = []
    for i in range(len(matrix)):
        row = []
        for j in range(len(matrix)):
            row.append(cofactor(matrix, i, j))
        adjoint.append(row)
    adjoint = transpose(adjoint)
    inverse = [[adjoint[i][j] / det for j in range(len(matrix))] for i in range(len(matrix))]
    return inverse

# Matriks yang diberikan
A = [
    [3, 1, 3, 4, 5],
    [2, 1, 2, 6, 7],
    [4, 2, 1, 8, 9],
    [5, 3, 7, 1, 0],
    [1, 2, 3, 4, 5]
]

# Menghitung determinan matriks
det_A = determinant(A)
print(f"Determinan dari matriks A adalah: {det_A}")

# Menghitung invers matriks
try:
    inverse_A = matrix_inverse(A)
    print("Inverse dari matriks A adalah:")
    for row in inverse_A:
        print(row)
except ValueError as e:
    print(e)
```
