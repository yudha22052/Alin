### **jenis jenis matrix tranformasi**


Transformasi matriks merupakan teknik matematika yang digunakan untuk mengubah bentuk atau posisi objek dalam ruang. ada beberapa jenis transformasi matriks yang umum digunakan dalam berbagai bidang, seperti grafika komputer, visi komputer, dan geometri:

##1. Rotasi (Rotation)
Rotasi memutar objek di sekitar titik asal (origin) dengan sudut tertentu. Matriks rotasi untuk rotasi 2D adalah:

$$
\begin{align*}
\text{Rumus}\\
 P' = R(0).P\\
 \text{Atau}
\end{align*}
$$

Rotasi 0°:

$$
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}
$$

Rotasi θ:

$$
\begin{bmatrix}
cos(𝜃)&-sin(𝜃)\\
sin(𝜃)&cos(𝜃)
\end{bmatrix}
$$


##2. Skala (Scaling)
Skala mengubah ukuran objek, memperbesar atau memperkecil objek relatif terhadap titik asal. Matriks skala untuk skala 2D adalah:

$$
\begin{align*}
\text{Rumus}\\
 P' = S(s_x , s_y ).P\\
 \text{Atau}
\end{align*}
$$

$$
\begin{bmatrix}
𝑠𝑥&0\\
0&𝑠𝑦
\end{bmatrix}
$$

di mana 𝑠𝑥 dan 𝑠𝑦 adalah faktor skala di sepanjang sumbu x dan y.

##3. Translasi (Translation)
Translasi menggeser objek dari satu lokasi ke lokasi lain. Translasi dalam koordinat homogen (dengan tambahan dimensi) untuk 2D adalah:

$$
\begin{align*}
\text{rumus}\\
 P' = T(t_x , t_y ).P\\
 \text{Atau}
\end{align*}
$$

$$
\begin{bmatrix}
1&0&𝑡𝑥\\
0&1&𝑡𝑦\\
0&0&1
\end{bmatrix}
$$

di mana 𝑡𝑥 dan 𝑡𝑦 adalah pergeseran di sepanjang sumbu x dan y.

##4. Refleksi (Reflection)
Refleksi mencerminkan objek terhadap garis atau bidang tertentu. Matriks refleksi untuk 2D adalah:

Refleksi terhadap sumbu x:

$$
\begin{bmatrix}
1&0\\
0&-1
\end{bmatrix}
$$

Refleksi terhadap sumbu y:

$$
\begin{bmatrix}
-1&0\\
0&1
\end{bmatrix}
$$

Refleksi terhadap garis y=x:

$$
\begin{bmatrix}
1&0\\
0&1
\end{bmatrix}
$$

##5. Shear (Geseran)
Shear menggeser satu bagian dari objek lebih dari bagian lain, menghasilkan efek miring. Matriks shear untuk 2D adalah:

Shear horizontal:

$$
\begin{bmatrix}
1&𝑘\\
0&1
\end{bmatrix}
$$

Shear vertikal:

$$
\begin{bmatrix}
1&0\\
𝑘&1
\end{bmatrix}
$$
di mana 𝑘 adalah faktor shear.

##  contoh code refleksi dan rotasi

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisi fungsi transformasi T
def T(x):
    return np.array([2 * x[0] + x[2], -4 * x[1]])

# Definisikan beberapa titik di ruang R^3
points_R3 = np.array([
    [1, 2, 3],
    [2, -1, 1],
    [-1, 0, 2],
    [0, -2, 3],
    [3, 1, -1]
])

# Terapkan transformasi pada setiap titik
points_R2 = np.array([T(point) for point in points_R3])

# Visualisasikan titik-titik sebelum dan sesudah transformasi
fig = plt.figure(figsize=(12, 6))

# Plot di R^3
ax = fig.add_subplot(121, projection='3d')
ax.scatter(points_R3[:, 0], points_R3[:, 1], points_R3[:, 2], color='b', label='Original Points (R^3)')
for point in points_R3:
    ax.text(point[0], point[1], point[2], f'({point[0]}, {point[1]}, {point[2]})')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('X3')
ax.set_title('Points in R^3')
ax.legend()
ax.grid(True)

# Plot di R^2
ax2 = fig.add_subplot(122)
ax2.scatter(points_R2[:, 0], points_R2[:, 1], color='r', label='Transformed Points (R^2)')
for point in points_R2:
    ax2.text(point[0], point[1], f'({point[0]}, {point[1]})')

ax2.axhline(0, color='grey', lw=0.5)
ax2.axvline(0, color='grey', lw=0.5)
ax2.set_xlabel('2X1 + X3')
ax2.set_ylabel('-4X2')
ax2.set_title('Points in R^2')
ax2.legend()
ax2.grid(True)

plt.show()
```

##  skala (scaling)
Rumus manual untuk melakukan scaling pada sebuah titik

$$
\begin{align*}
\text{Scaled point} &= (s_x \times x, s_y \times y)
\end{align*}
$$

Dalam bentuk matriks, kita bisa menuliskan operasi ini sebagai berikut:

$$
\begin{bmatrix}
s_x & 0 \\
0 & s_y
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
$$

Di mana (𝑥, 𝑦) adalah koordinat titik sebelum scaling, dan matriks skala adalah matriks diagonal dengan 𝑠𝑥 dan 𝑠𝑦 sebagai elemen diagonal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisi fungsi untuk melakukan scaling menggunakan matriks transformasi affine
def scale(points, scale_factors):
    # Matriks transformasi affine untuk scaling
    transformation_matrix = np.array([
        [scale_factors[0], 0, 0],
        [0, scale_factors[1], 0],
        [0, 0, 1]
    ])

    # Tambahkan kolom ones untuk koordinat homogen
    homogenous_coordinates = np.hstack([points, np.ones((points.shape[0], 1))])

    # Melakukan perkalian matriks untuk menerapkan transformasi
    scaled_points_homogeneous = np.dot(homogenous_coordinates, transformation_matrix.T)

    # Kembalikan hasil setelah mengubah koordinat homogen kembali ke format biasa
    scaled_points = scaled_points_homogeneous[:, :-1]

    return scaled_points

# Definisikan titik-titik awal
points = np.array([
    [1, 1],
    [2, 3],
    [3, 2],
    [4, 5]
])

# Faktor skala untuk masing-masing sumbu
scale_factors = [2, 3]  # Faktor skala untuk sumbu X dan Y

# Terapkan scaling pada setiap titik menggunakan matriks transformasi affine
scaled_points = scale(points, scale_factors)

# Visualisasikan titik-titik sebelum dan sesudah scaling
plt.figure(figsize=(8, 4))

# Plot titik-titik awal
plt.scatter(points[:, 0], points[:, 1], color='b', label='Original Points')
for i, point in enumerate(points):
    plt.text(point[0], point[1], f'P{i+1}')

# Plot titik-titik setelah scaling
plt.scatter(scaled_points[:, 0], scaled_points[:, 1], color='r', label='Scaled Points')
for i, point in enumerate(scaled_points):
    plt.text(point[0], point[1], f'S{i+1}')

# Penambahan label dan judul
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scaling of Points using 3x3 Transformation Matrix')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```

## 5. Shear (Geseran)

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat matriks shear horizontal
def shear_horizontal(k, points):
    shear_matrix = np.array([
        [1, k],
        [0, 1]
    ])
    return np.dot(shear_matrix, points)

# Buat grid titik-titik
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x, y)
points = np.vstack([X.ravel(), Y.ravel()])

# Shear horizontal
k_horizontal = 0.5
sheared_points_horizontal = shear_horizontal(k_horizontal, points)

# Plot
plt.figure(figsize=(8, 6))
plt.title('Shear Horizontal')
plt.plot(points[0], points[1], color='blue', linestyle='-', label='Original Points')
plt.plot(sheared_points_horizontal[0], sheared_points_horizontal[1], color='red', linestyle='-', label='Sheared Points')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()
```
