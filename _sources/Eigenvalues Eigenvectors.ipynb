```python
import numpy as np

# Define the matrix A
A = np.array([[1, 4],
              [2, 3]])

# Find the eigenvalues of A
eigenvalues = np.linalg.eigvals(A)

print("The eigenvalues of matrix A are:", eigenvalues)
```

```python
import numpy as np
import sympy as sp

# Define the matrix A
A = np.array([[1, 4],
              [2, 3]])

# Define the symbol λ
lmbd = sp.symbols('λ')

# Define the identity matrix
I = np.eye(2)

# Convert numpy matrices to sympy matrices
A_sympy = sp.Matrix(A)
I_sympy = sp.Matrix(I)

# Calculate A - λI
A_minus_lmbdI = A_sympy - lmbd * I_sympy

# Calculate the determinant of A - λI
det_A_minus_lmbdI = A_minus_lmbdI.det()

# Print the characteristic polynomial
print(f"The characteristic polynomial is: det(A - λI) = {det_A_minus_lmbdI}")

# Solve the characteristic polynomial for λ (eigenvalues)
eigenvalues = sp.solve(det_A_minus_lmbdI, lmbd)

print("The eigenvalues of matrix A are:", eigenvalues)
```

```python
import numpy as np

A = np.array([[1, 4],
              [2, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("matrix A:\n",A)
print("Eigenvalues matrix A:", eigenvalues)
print("Eigenvectors matrix A:\n", eigenvectors)
```

```python
import numpy as np

# Matriks yang diberikan
A = np.array([[1, 4],
              [2, 3]])

# Menghitung eigenvalues
eigenvalues, _ = np.linalg.eig(A)
print("Eigenvalues matrix A:", eigenvalues)

# Menghitung λ-1 untuk setiap eigenvalue
eigenvalue_minus_one = eigenvalues - 1
print("λ-1:", eigenvalue_minus_one)
```
