##merubah baris akan merubah tanda dan hasil dari deeterminan

```python
import numpy as np
matrix = np.array([[1, 2, 1], [3, 1, 2], [2, 1, 2]])
matrix1 = np.array([[3, 1, 2], [1, 2, 1], [2, 1, 2]])
c = round(np.linalg.det(matrix)) #bilbul
d = round(np.linalg.det(matrix1))
print('sebelum di ubah',c)
print('setelah di ubah',d)
```

##jika sebuah baris merupakan kelipatan dari baris yang lain maka matrix akan memiliki determinan 0

```python
import numpy as np
matrix = np.array([[1, 2, 1], [2, 4, 2], [2, 1, 2]])
matrix1 = np.array([[2, 4, 2], [1, 2, 1], [2, 1, 2]])
e = round(np.linalg.det(matrix)) #bilbul
determinant1 = round(np.linalg.det(matrix1))
print('baris 2 perkalian baris 1 : ',e)
print('baris 2 di tukar dengan baris 1 : ',determinant1)
```

###untuk mengetahui kesamaan matrik bi di hitung dengan kelipatan dipangkat dengan ordo metrik dan di kali dengan hasil

```python
import numpy as np
matrix = np.array([[3, 4], [2, 1]])
matrix1 = np.array([[6, 8], [4, 2]])
A = round(np.linalg.det(matrix))
h = round(np.linalg.det(matrix1))
print(A)
print(h)
```

###apabila salah satu baris A dikalikan 2 makan hasil kali determinan matrix i samadengan determinan matrix i di kali 2

##determinan dari segitiha atas dan segitiga bawah

```python
import numpy as np
matrix = np.array([[2, 4, 2], [0, 3, 2], [0, 0, 1]])
matrix1 = np.array([[1, 0, 0], [2, 3, 0], [2, 4, 2]])
determinant1 = round(np.linalg.det(matrix)) #bilbul
determinant2 = round(np.linalg.det(matrix1))
print('membuat matrix segitiga atas : ',determinant1)
print('membuat matrix segitiga bawah : ',determinant2)
```

##perkalian dua matrix

\begin{align*}
B = CD
&&
C=
\begin{bmatrix}
2&1\\
1&1
\end{bmatrix}
&&
D=
\begin{bmatrix}
1&1\\
1&1
\end{bmatrix}
\end{align*}

```python
import numpy as np
matrix = np.array([[2, 1], [1, 1]])
matrix1 = np.array([[1, 1], [1, 1]])
C = round(np.linalg.det(matrix))
D = round(np.linalg.det(matrix1))
print('determinan matrix C : ',C)
print('determinan matrix D : ',D)
```

###hasilnya

```python
import numpy as np
matrix = np.array([[3, 3], [2, 2]])
B = round(np.linalg.det(matrix))
print('hasil dari perkalian 2 determinan matrix : ',B)
```
